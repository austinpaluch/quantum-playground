# Quantum-Inspired Optimization
Using QAOA for a simple sales portfolio optimization (toy example: maximize profit with constraints).

```python
from qiskit import QuantumCircuit
from qiskit_algorithms import QAOA
from qiskit_algorithms.optimizers import COBYLA
from qiskit_aer import AerSimulator
from qiskit_optimization import QuadraticProgram
from qiskit_optimization.algorithms import MinimumEigenOptimizer
from qiskit.primitives import StatevectorSampler as Sampler
import numpy as np
```

## Toy Sales Problem
Maximize: 3x + 2y (profit)
Subject to: x + y <= 1 (budget)
x,y binary.

```python
qp = QuadraticProgram()
qp.binary_var('x')
qp.binary_var('y')
qp.maximize(linear={'x':3, 'y':2})
qp.linear_constraint({'x':1, 'y':1}, '<=', 1)
print(qp.prettyprint())
```

```python
optimizer = COBYLA()
sampler = Sampler()
qaoa = QAOA(sampler, optimizer, reps=1)
meo = MinimumEigenOptimizer(qaoa)
result = meo.solve(qp)
print(result)
```
