# Basic Quantum Circuits
This notebook covers Hello World, superposition, and Bell states using Qiskit.

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

## Hello World: Single Qubit Circuit
Create a simple circuit with one qubit and measure it.

```python
# Create a Quantum Circuit with 1 qubit and 1 classical bit
qc = QuantumCircuit(1, 1)

# Measure the qubit
qc.measure(0, 0)

# Draw the circuit
qc.draw('mpl')
```

```python
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
result = simulator.run(compiled_circuit, shots=1024).result()
counts = result.get_counts()
plot_histogram(counts)
```

## Superposition with Hadamard Gate
Apply Hadamard to create superposition.

```python
qc_super = QuantumCircuit(1, 1)
qc_super.h(0)  # Hadamard gate on qubit 0
qc_super.measure(0, 0)
qc_super.draw('mpl')
```

## Bell State: Entanglement
Create a Bell state with two qubits.

```python
qc_bell = QuantumCircuit(2, 2)
qc_bell.h(0)  # Hadamard on qubit 0
qc_bell.cx(0, 1)  # CNOT with control 0, target 1
qc_bell.measure([0, 1], [0, 1])
qc_bell.draw('mpl')
```

```python
compiled_bell = transpile(qc_bell, simulator)
result_bell = simulator.run(compiled_bell, shots=1024).result()
counts_bell = result_bell.get_counts()
plot_histogram(counts_bell)
```
